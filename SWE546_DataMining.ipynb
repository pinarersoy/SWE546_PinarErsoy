```python
# Bar Plot

import numpy as np
import matplotlib.pyplot as plt

n = 10
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

plt.ylim(-1.25,+1.25)
plt.show()
```

```python
# Pie Chart

import numpy as np
import matplotlib.pyplot as plt

n = 10
Z = np.random.uniform(0,1,n)
plt.pie(Z)
plt.show()
```

```python

# HW_2 
# Scatter Plot & Heatmap examples

# Scatter Plot

from matplotlib import pyplot as plt
import numpy as np
import matplotlib

x = np.arange(18.0, 50.0, 2.0) # Bank can start to give credit to customers that are above 18 years old
y = -x**2.1 + np.random.randn(*x.shape) * 30.00

s = np.random.rand(*x.shape) * 300 + 100

plt.scatter(x, y, s, c="g", alpha=0.8, marker=r'$\clubsuit$',label="Garanti")

plt.xlabel("Customer Age")
plt.ylabel("Credit Risk")

plt.legend(loc=1) #locate the logo in the 1st quarter
plt.show()
```

```python

# Heatmap Example
import numpy as np
import numpy.random
import matplotlib.pyplot as plt

# Random data generation 
x = np.random.randn(10000)
y = np.random.randn(10000)

heatmap, xedges, yedges = np.histogram2d(x, y, bins=30)

# Locate the edges in the given periods
extent = [xedges[0], xedges[-2], yedges[0], yedges[-2]]

plt.clf()
plt.imshow(heatmap.T, extent=extent)
plt.show()
```
